```python
from copy import *
class Sequence():
    def __init__(self, l):
        self.l = list(l)
        
    def __str__(self):
        return str(self.l)
        
    def __eq__(self,other):
        return self.l==other.l
    
    def __getitem__(self,n):
        return self.l[n]
    
    def __delitem__(self, key):
        del self.l[key]
        
    def __iter__(self):
        return iter(self.l)
    
    def insert(self,position,value):
        return self.l.insert(position,value)
    
    def append(self,value):
        return self.l.append(value)
        
    def __copy__(self):
        return copy.deepcopy(self)
    
        
        
    
    
    
```

```python

s = Sequence(range(12))
s1 = Sequence(range(12))
s2 = Sequence(range(10))
#print(s)

s.append(999)
s = deepcopy(s2)
s is s2
print(s2)
```
